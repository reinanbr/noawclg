```python
import numpy as np
from scipy.interpolate import interp1d

def interpolate_temporal_data(data, factor=3):
    """
    Interpolates data along the temporal dimension.
    
    Parameters:
    -----------
    data : numpy.ndarray
        Array with shape (time, lat, lon) - e.g., (129, 721, 1440)
    factor : int
        Multiplication factor (3 = triple the temporal resolution)
    
    Returns:
    --------
    numpy.ndarray
        Interpolated array with shape (time*factor, lat, lon) - e.g., (387, 721, 1440)
    """
    n_time_original, n_lat, n_lon = data.shape
    
    # Create original and new indices
    original_indices = np.linspace(0, n_time_original - 1, n_time_original)
    n_time_new = (n_time_original - 1) * factor + 1  # 387 for factor=3
    new_indices = np.linspace(0, n_time_original - 1, n_time_new)
    
    # Array to store result
    interpolated_data = np.zeros((n_time_new, n_lat, n_lon), dtype=data.dtype)
    
    # Interpolate for each grid point (lat, lon)
    print(f"Interpolating from {n_time_original} to {n_time_new} timesteps...")
    
    for i in range(n_lat):
        if i % 100 == 0:
            print(f"Processing latitude {i}/{n_lat}")
        
        for j in range(n_lon):
            # Time series for this point
            time_series = data[:, i, j]
            
            # Create linear interpolator
            interpolator = interp1d(original_indices, time_series, 
                                   kind='linear', fill_value='extrapolate')
            
            # Apply interpolation
            interpolated_data[:, i, j] = interpolator(new_indices)
    
    print(f"Interpolation complete! Final shape: {interpolated_data.shape}")
    return interpolated_data


# Usage example
if __name__ == "__main__":
    # Simulate example data
    print("Creating example data...")
    example_data = np.random.randn(129, 721, 1440).astype(np.float32)
    
    print(f"Original shape: {example_data.shape}")
    
    # Interpolate with factor 3
    interpolated_data = interpolate_temporal_data(example_data, factor=3)
    
    print(f"\nInterpolated shape: {interpolated_data.shape}")
    print(f"Expected: (387, 721, 1440)")
    
    # Check temporal continuity
    print("\nVerifying interpolation (first values):")
    print(f"Original[0]: {example_data[0, 360, 720]:.4f}")
    print(f"Interpolated[0]: {interpolated_data[0, 360, 720]:.4f}")
    print(f"Interpolated[1]: {interpolated_data[1, 360, 720]:.4f}")
    print(f"Interpolated[2]: {interpolated_data[2, 360, 720]:.4f}")
    print(f"Interpolated[3]: {interpolated_data[3, 360, 720]:.4f}")
    print(f"Original[1]: {example_data[1, 360, 720]:.4f}")
```

```python

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from animateplot import AnimatePlot
from noawclg import get_noaa_data as gnd
from IPython.display import display

from datetime import datetime
from datetime import datetime, timedelta

def generate_timestamps(quantity=128, interval_hours=3, start_date=None):
    """
    Generates a list of timestamps with specified interval.
    
    Args:
        quantity (int): Number of timestamps to generate (default: 128)
        interval_hours (int): Interval in hours between each timestamp (default: 3)
        start_date (datetime): Initial date/time (default: now)
    
    Returns:
        list: List of timestamps in datetime format
    """
    if start_date is None:
        start_date = datetime.now()
    
    timestamps = []
    for i in range(quantity):
        timestamp = start_date + timedelta(hours=interval_hours * i)
        timestamps.append(timestamp)
    
    return timestamps








timestamps = generate_timestamps(quantity=128, interval_hours=3, start_date=datetime(2025, 10, 19, 0, 0, 0))

data_noaa = gnd(date='23/10/2025')
wind = data_noaa['ugrdmwl'].to_numpy()

def plot_wind(i, plt):
    data = wind[i]

    plt.figure(figsize=(14, 10), facecolor='black')

    # Cria o mapa
    m = Basemap(
        projection='ortho',
        resolution='l',
        lat_0=-9, lon_0=-40
    )
    
    m.drawcoastlines(linewidth=0.5)
    m.drawcountries(linewidth=0.3)
    m.fillcontinents(color='lightgray', lake_color='white')
    m.drawmapboundary(fill_color='white')
    
    # Meridianos sem labels
    m.drawmeridians(np.arange(-180, 181, 60), linewidth=0.5, color='gray', alpha=0.5)
    
    

    # Linhas importantes SEM labels
    # Equador
    m.drawparallels([0], linewidth=2, color='red', dashes=[1,0])
    
    # Trópicos
    m.drawparallels([23.5, -23.5], linewidth=.5, color='orange', dashes=[5,3])
    
    # Círculos Polares
    m.drawparallels([66.5, -66.5], linewidth=.5, color='blue', dashes=[5,3])
    
    # Paralelos adicionais
    m.drawparallels(np.arange(-90, 91, 30), linewidth=0.1, color='gray', alpha=0.5)
    
    lats = np.linspace(90, -90, 721)
    lons = np.linspace(-180, 180, 1440)
    
    lon2d, lat2d = np.meshgrid(lons, lats)
    x, y = m(lon2d, lat2d)
    
    # Define os limites fixos baseados em todos os dados
    vmin = wind.min()
    vmax = wind.max()

    

    cs_ = m.contourf(x, y, wind[0], levels=20, cmap='magma', extend='both', 
                    vmin=vmin, vmax=vmax)
    cs = m.contourf(x, y, data, levels=20, cmap='magma', extend='both', 
                    vmin=vmin, vmax=vmax)
    # Colorbar menor e mais compacta
    cbar = plt.colorbar(cs_, orientation='horizontal', pad=0.03, shrink=0.5, aspect=30)
    cbar.set_label('Velocidade do Vento (m/s)', fontsize=9, color='white')
    cbar.ax.tick_params(labelsize=8, colors='white')

    plt.title(f'Jatos de Vento \n {timestamps[i].strftime("%d %B %Y %H:%M")}', fontsize=14, fontweight='bold', color='white')

    return plt


animator = AnimatePlot(
    space=list(range(128)),
    callplot=plot_wind,
    dpi=120
)

# 2. Verifica se há cache
cache_info = animator.get_cache_info()
if cache_info['exists']:
    print(f"Cache encontrado: {cache_info['count']} imagens")
    # Limpa cache antigo
    animator.delete_cache()

# 3. Renderiza novos frames
animator.render_cache(show_progress=True)

# 4. Gera outputs
animator.render_gif('output_ortho.gif', fps=15)
video_widget = animator.render_mp4('output_ortho.mp4', fps=15, use_opencv=False)

# 5. Se estiver no Jupyter, o vídeo será exibido automaticamente
if video_widget:
    display(video_widget)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.animation import FuncAnimation, FFMpegWriter
from IPython.display import HTML

def create_temporal_video(data_array, output_filename='temporal_animation.mp4', 
                         fps=10, title_prefix='Timestep', cmap='jet',
                         vmin=None, vmax=None):
    """
    Create a temporal video animation from 3D array data.
    
    Parameters:
    -----------
    data_array : numpy.ndarray
        Array with shape (time, lat, lon) - e.g., (129, 721, 1440) or (387, 721, 1440)
    output_filename : str
        Name of the output video file
    fps : int
        Frames per second for the video
    title_prefix : str
        Prefix for the title shown in each frame
    cmap : str
        Colormap name
    vmin, vmax : float
        Min and max values for color scale (None = auto)
    
    Returns:
    --------
    animation : matplotlib.animation.FuncAnimation
        The animation object
    """
    n_timesteps, n_lat, n_lon = data_array.shape
    
    # Create latitude and longitude grids
    lats = np.linspace(90, -90, n_lat)
    lons = np.linspace(0, 360, n_lon)
    lon2d, lat2d = np.meshgrid(lons, lats)
    
    # Set color scale limits if not provided
    if vmin is None:
        vmin = np.nanmin(data_array)
    if vmax is None:
        vmax = np.nanmax(data_array)
    
    print(f"Creating animation with {n_timesteps} frames...")
    print(f"Data range: [{vmin:.2f}, {vmax:.2f}]")
    
    # Create figure and basemap
    fig = plt.figure(figsize=(12, 6))
    m = Basemap(projection='cyl', resolution='c', 
                llcrnrlat=90, urcrnrlat=-90,
                llcrnrlon=0, urcrnrlon=360)
    
    m.drawparallels(np.arange(-90, 91, 30), labels=[1,0,0,0])
    m.drawmeridians(np.arange(0, 361, 60), labels=[0,0,0,1])
    m.drawcoastlines()
    
    # Initial plot
    cs = m.pcolormesh(lon2d, lat2d, data_array[0], latlon=True, 
                      shading='auto', cmap=cmap, vmin=vmin, vmax=vmax)
    
    cbar = plt.colorbar(cs, orientation='vertical', label='Value')
    title = plt.title(f'{title_prefix}: 0 / {n_timesteps-1}')
    
    plt.tight_layout()
    
    def update(frame):
        """Update function for animation"""
        cs.set_array(data_array[frame].ravel())
        title.set_text(f'{title_prefix}: {frame} / {n_timesteps-1}')
        
        if frame % 10 == 0:
            print(f"Processing frame {frame}/{n_timesteps-1}")
        
        return cs, title
    
    # Create animation
    anim = FuncAnimation(fig, update, frames=n_timesteps, 
                        interval=1000/fps, blit=False, repeat=True)
    
    # Save animation
    print(f"\nSaving video to {output_filename}...")
    writer = FFMpegWriter(fps=fps, bitrate=2000)
    anim.save(output_filename, writer=writer, dpi=100)
    print(f"Video saved successfully!")
    
    return anim


def create_temporal_video_inline(data_array, fps=10, title_prefix='Timestep', 
                                 cmap='jet', vmin=None, vmax=None):
    """
    Create a temporal video animation for Jupyter notebooks (inline display).
    
    Parameters: Same as create_temporal_video
    
    Returns:
    --------
    HTML : IPython.display.HTML
        HTML5 video player for notebook
    """
    n_timesteps, n_lat, n_lon = data_array.shape
    
    # Create latitude and longitude grids
    lats = np.linspace(90, -90, 721)       # latitude decrescendo
    lons = np.linspace(0, 360, 1440)       # longitude de 0 a 360

    lon2d, lat2d = np.meshgrid(lons, lats)

    # Set color scale limits
    if vmin is None:
        vmin = np.nanmin(data_array)
    if vmax is None:
        vmax = np.nanmax(data_array)
    
    print(f"Creating animation with {n_timesteps} frames...")
    
    # Create figure and basemap
    fig = plt.figure(figsize=(12, 6))
    m = Basemap(projection='cyl', resolution='c', 
                llcrnrlat=90, urcrnrlat=-90,
                llcrnrlon=0, urcrnrlon=360)
    
    m.drawparallels(np.arange(-90, 91, 30), labels=[1,0,0,0])
    m.drawmeridians(np.arange(0, 361, 60), labels=[0,0,0,1])
    m.drawcoastlines()
    
    # Initial plot
    cs = m.pcolormesh(lon2d, lat2d, data_array[0], latlon=True, 
                      shading='auto', cmap=cmap, vmin=vmin, vmax=vmax)
    
    cbar = plt.colorbar(cs, orientation='vertical', label='Value')
    title = plt.title(f'{title_prefix}: 0 / {n_timesteps-1}')
    
    plt.tight_layout()
    
    def update(frame):
        """Update function for animation"""
        cs.set_array(data_array[frame].ravel())
        title.set_text(f'{title_prefix}: {frame} / {n_timesteps-1}')
        return cs, title
    
    # Create animation
    anim = FuncAnimation(fig, update, frames=n_timesteps, 
                        interval=1000/fps, blit=False, repeat=True)
    
    plt.close()
    
    return HTML(anim.to_html5_video())


# Example usage
if __name__ == "__main__":
    # Simulate data (replace with your tmp_np array)
    print("Creating example data...")
    tmp_np = wind
    
    # Option 1: Save to file
    create_temporal_video(tmp_np, 
                         output_filename='temperature_evolution.mp4',
                         fps=15,
                         title_prefix='Temperature - Frame',
                         cmap='jet')
    
    # Option 2: Display inline in Jupyter (uncomment if using notebook)
    # video = create_temporal_video_inline(tmp_np, fps=15, 
    #                                       title_prefix='Temperature - Frame',
    #                                       cmap='jet')
    # display(video)
```
