```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Dados reais
data = tmp_np[0]

# Latitudes e longitudes
lats = np.linspace(90, -90, 721)
lons = np.linspace(0, 360, 1440)
lon2d, lat2d = np.meshgrid(lons, lats)

# Mapa ortográfico
plt.figure(figsize=(10, 10))
m = Basemap(projection='ortho', lat_0=0, lon_0=0, resolution='l')
m.drawcoastlines()
m.drawparallels(np.arange(-90, 91, 30), labels=[1,0,0,0], fontsize=8)
m.drawmeridians(np.arange(0, 361, 60), labels=[0,0,0,1], fontsize=8)

# Coordenadas projetadas
x, y = m(lon2d, lat2d)

# Máscara para fora da projeção
mask = np.isnan(x) | np.isnan(y)
x_masked = np.ma.masked_where(mask, x)
y_masked = np.ma.masked_where(mask, y)
data_masked = np.ma.masked_where(mask, data)

# Níveis personalizados para contornos (ajuste conforme o intervalo dos dados)
vmin = np.nanmin(data)
vmax = np.nanmax(data)
levels = np.linspace(vmin, vmax, 30)  # 30 níveis

# Contourf com mais detalhes
cs = m.contourf(x_masked, y_masked, data_masked, levels=levels, cmap='inferno', extend='both')

# Contornos opcionais para destaque
m.contour(x_masked, y_masked, data_masked, levels=levels, colors='k', linewidths=0.3, alpha=0.3)

# Barra de cores
plt.colorbar(cs, orientation='horizontal', pad=0.05, label='Temperatura (K)')
m.shadedrelief()

plt.title('Mapa do Globo com Contornos Refinados (Ortográfico)')
plt.tight_layout()
plt.show()
```
